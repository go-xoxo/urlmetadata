"""Simple command-line example for Custom Search.

Command-line application that does a search.
"""

```python
__author__ = "jcgregorio@google.com (Joe Gregorio)"

import pprint

from googleapiclient.discovery import build
```

```python
def main():
    # Build a service object for interacting with the API. Visit
    # the Google APIs Console <http://code.google.com/apis/console>
    # to get an API key for your own application.
    service = build(
        "customsearch", "v1", developerKey="AIzaSyChRaV9dvScw843jQdqCtQYvdGW9-gOK2k"
    )

    res = (
        service.cse()
        .list(
            q="lectures", #query
            cx="b7fe3192b01e947fe",
            sort="date:r", #sort by date
            num=10, #number of results to return
            hl="en", #interface language
            orTerms="", #additional search terms
            filter="1", #1=duplicates
            lr="", #lang_en for english only documents
            dateRestrict="m[12]",
            safe="off", #safe browsing filter
            hq="", #appends terms to query
            exactTerms="", #appends query terms
            excludeTerms="", #excludes query terms
            #linkSite="", #restricts results to URLs from a specified site
            #siteSearch="", #specifies all search results should be pages from a given site
            #siteSearchFilter="", #specifies whether to (i)nclude or (e)xclude siteSearch
            fileType="", #pdf etc.
            gl="", #geolocation of user
            relatedSite="", #finds sites related to a given site
            #highRange="", #specifies the ending value for a search range
            #lowRange="", #specifies the starting value for a search range
            #searchType="", #specifies the search type: image, news, video
            #imgSize="", #specifies the image size
            #imgType="", #specifies the image type
            #rights="", #specifies the rights of the image
            cr="", #country restrict(s)
            #googlehost="", #specifies the Google domain (e.g., google.com, google.de, google.fr etc.)
            #start="", #specifies the index of the first result to return
        )
        .execute()
    )
    pprint.pprint(res)
```

```python
if __name__ == "__main__":
    main()
```
